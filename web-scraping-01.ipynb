# Extracting text from the tag in web page using python

```python
lines = soup.find_all("p")
```

```python
lines

# this all data of p tag coes in list
```

```python
for l in lines:
    print(l.text)
```

```python
s = soup.find("div",class_ = "row")
```

```python
s
```

```python
lines_1 = s.find_all("p")
lines_1
```

```python
for l1 in lines_1:
    print(l1.text)
```

```python
# OR
```

```python
s1 = soup.find("h3",class_ = "content-wrapper ps-2 align-middle d-flex align-items-center")
s1
```

```python
s1 = soup.find("h3", class_="fs-16 lh-30 fw-500 label-color-5 my-1 heading-turnicate")
print(s1.text)  
```

```python
s1 = soup.find("h3", class_="fs-16 lh-30 fw-500 label-color-5 my-1 heading-turnicate")
print(s1.text.strip())  

# strip()-->> used to remove extra space in pythoon from begnning to end
```

## Want to extract all such headings (like "SEO Quiz", "Java Programs", etc.)?

```python
soup.find_all("h3", class_="fs-16 lh-30 fw-500 label-color-5 my-1 heading-turnicate")
```

```python
headings = soup.find_all("h3", class_="fs-16 lh-30 fw-500 label-color-5 my-1 heading-turnicate")

for h in headings:
    print(h.text.strip())
```

#### h.text	✅	Single tag
h.text.strip()	✅	Cleaned text from one tag

headings.text	❌	headings is a list

### Extracting Links in Web Page using Python

```python
soup.find_all("a")

# anchor tag contains all the links
```

```python
for i in soup.find_all("a"):
    print(i)
```

```python
for i in soup.find_all("a"):
    print(i.get("href"))
```
